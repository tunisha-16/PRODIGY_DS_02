```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('C:/Users/KarangulaTunisha/Downloads/test.csv')

# Display the first few rows of the dataset
print(titanic_data.head())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())

# Summary statistics for numerical variables
print(titanic_data.describe())

# Summary statistics for categorical variables
print(titanic_data.describe(include=['O']))
```

```python
# Print all column names in the Titanic dataset
print(titanic_data.columns)
```

```python
# Summary statistics for categorical variables
categorical_columns = ['Sex', 'Embarked']
print(titanic_data[categorical_columns].describe())
```

```python
# Visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of passengers by sex
sns.countplot(x='Sex', data=titanic_data)
plt.title('Passenger Count by Sex')
plt.show()
# Distribution of passengers by passenger class
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Passenger Count by Passenger Class')
plt.show()

# Age distribution of passengers
sns.histplot(titanic_data['Age'], kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.show()

# Fare distribution
sns.histplot(titanic_data['Fare'], kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset('iris')

# Display the first few rows of the dataset
print(iris_data.head())

# Check for missing values
print(iris_data.isnull().sum())

# Distribution of flowers by species
sns.countplot(x='species', data=iris_data)
plt.title('Flower Count by Species')
plt.show()

# Distribution of sepal length
sns.histplot(iris_data['sepal_length'], kde=True)
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.show()

# Distribution of petal width by species
sns.boxplot(x='species', y='petal_width', data=iris_data)
plt.title('Petal Width Distribution by Species')
plt.show()
```
